# Date and time transformations

```python
# Determind if the transaction is occured during weekend
%time transactions_df['TX_DURING_WEEKEND']=transactions_df.TX_DATETIME.apply(utils_transform_data.is_weekend)
```

```python
# Determind if the transaction is ocurred during night
%time transactions_df['TX_DURING_NIGHT']=transactions_df.TX_DATETIME.apply(utils_transform_data.is_night)
```

```python
transactions_df[transactions_df.TX_TIME_DAYS>=30]
```

# Customer ID transformations

```python
spending_behaviour_customer_0=utils_transform_data.get_customer_spending_behaviour_features(transactions_df[transactions_df.CUSTOMER_ID==0])
spending_behaviour_customer_0
```

```python
# Analyse each customer's spending behavior base on the different day range's avg spending amount
%time transactions_df=transactions_df.groupby('CUSTOMER_ID').apply(lambda x: utils_transform_data.get_customer_spending_behaviour_features(x, windows_size_in_days=[1,7,30]))
transactions_df=transactions_df.sort_values('TX_DATETIME').reset_index(drop=True)
```
